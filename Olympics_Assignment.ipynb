```python
import numpy as np
import pandas as pd
```

# SUMMER OLYMPICS DATA ANALYSIS

### 1. In how many cities Summer Olympics is held so far?

```python
df = pd.read_csv("summer.csv") 
```

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold_medal = df[df['Medal'] == 'Gold']
 
lst = []
for i in df_gold_medal['Sport'].unique():
     lst.append([i, len(df_gold_medal[df_gold_medal['Sport'] == i])])
 
pd.DataFrame(lst, columns = ['Sport','Gold_Medals']).sort_values(by = 'Gold_Medals', ascending = False).head().plot.bar(x = 'Sport', y = 'Gold_Medals')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
lsts = []
for i in df['Sport'].unique():
     lsts.append([i, len(df[df['Sport'] == i])])
 
pd.DataFrame(lsts, columns = ['Sport','Medals']).sort_values(by = 'Medals', ascending = False).head().plot.bar(x = 'Sport', y = 'Medals')
```

### 4. Which player has won most number of medals? (Top 5)


```python
lstss = []
for i in df['Athlete'].unique():
     lstss.append([i, len(df[df['Athlete'] == i])])
 
pd.DataFrame(lstss, columns = ['Player', 'Medals']).sort_values(by = 'Medals', ascending = False).head().plot.bar(x = 'Player' , y = 'Medals')
```

### 5. Which player has won most number of Gold Medals? (Top 5)

```python
df_gold_medals = df[df['Medal'] == 'Gold']
 
lstsss = []
for i in df_gold_medals['Athlete'].unique():
     lstsss.append([i, len(df_gold_medals[df_gold_medals['Athlete'] == i])])
 
pd.DataFrame(lstsss, columns = ['Player','Gold_Medals']).sort_values(by = 'Gold_Medals', ascending = False).head().plot.bar(x = 'Player' , y = 'Gold_Medals')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df_Gold_Medal = df[df['Medal'] == 'Gold']
df_Gold_India = df_Gold_Medal[df_Gold_Medal['Country'] == 'IND']
 
lstt = []
for i in df_Gold_India['Year'].unique():
     lstt.append(i)
min(lstt)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
lstssss = []
for i in df['Event'].unique():
     lstssss.append([i, len(df[df['Event'] == i])])
 
pd.DataFrame(lstssss,columns = ['Event','Total_Players']).sort_values(by = 'Total_Players', ascending = False).head().plot.bar(x = 'Event', y = 'Total_Players')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df_gold_medalss = df[df['Medal'] == 'Gold']
df_gold_women = df_gold_medalss[df_gold_medalss['Gender'] == 'Women']
 
lstsssss = []
for i in df_gold_women['Sport'].unique():
     lstsssss.append([i,len(df_gold_women[df_gold_women['Sport'] == i])])
 
pd.DataFrame(lstsssss, columns = ['Sport', 'Gold_Medals_For_Female']).sort_values(by = 'Gold_Medals_For_Female', ascending = False).head().plot.bar(x = 'Sport', y = 'Gold_Medals_For_Female')
```
